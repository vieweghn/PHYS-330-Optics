```python
import numpy as np
import matplotlib.pyplot as plt

D = 10e-6
lam = 500e-9
k = 2 * np.pi / lam
I0 = 1.0

z = np.logspace(-6, -3, 1000)

phase_exact = k * (np.sqrt(z**2 + (D/2)**2) - z)
I_exact = 2 * I0 * (1 - np.cos(phase_exact))

# Fresnel Approximation
phase_fresnel = (k * D**2) / (16 * z)
I_fresnel = 4 * I0 * np.sin(phase_fresnel)**2

#  Fraunhofer Approximation
beta = (k * D**2) / (8 * z)
I_fraunhofer = I0 * beta**2


plt.figure(figsize=(10, 6))
plt.plot(z, I_exact, label='Exact (Huygens)', color='black', linewidth=3)
plt.plot(z, I_fresnel, label='Fresnel Approx.', color='red', linestyle='--', linewidth=2)
plt.plot(z, I_fraunhofer, label='Fraunhofer Approx.', color='blue', linestyle=':', linewidth=2)

plt.xscale('log')
plt.ylim(0, 5) # Cap Y axis to see oscillations clearly
plt.xlabel('Distance z (meters) - Log Scale')
plt.ylabel('On-Axis Intensity (I/I0)')
plt.legend()
plt.grid()
plt.show()
```
