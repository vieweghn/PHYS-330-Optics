```python
import sympy as sp
```

```python
u_x, u_y, u_z = sp.symbols(r"u_x, u_y, u_z", real=True)
n_x, n_y, n_z = sp.symbols(r"n_x, n_y, n_z", real=True)
n = sp.symbols("n")
subsVals = {u_x:1/sp.sqrt(3), u_y:1/sp.sqrt(3), u_z:1/sp.sqrt(3), n_x:sp.Rational(15,10), n_y:sp.Rational(18,10), n_z:sp.Rational(20,10)}
```

```python
one_over_n_squared = sp.symbols(r"1/n^2", real=True)

fresnel_equation = (u_x**2 / (one_over_n_squared - 1/n_x**2) +
                    u_y**2 / (one_over_n_squared - 1/n_y**2) +
                    u_z**2 / (one_over_n_squared - 1/n_z**2))

fresnel_equation_substituted = fresnel_equation.subs({u_x: subsVals[u_x], u_y: subsVals[u_y], u_z: subsVals[u_z]})
print(fresnel_equation_substituted)
```

```python
solns_one_over_n_squared = sp.solve(fresnel_equation_substituted, one_over_n_squared)
print("Solutions for 1/n^2:")
print(solns_one_over_n_squared)
```

```python
numerical_solns_one_over_n_squared = [sol.subs(subsVals) for sol in solns_one_over_n_squared]

# Solve for n
refractive_indices = [sp.sqrt(1/sol) for sol in numerical_solns_one_over_n_squared]

print("\nCalculated refractive indices (n):")
for idx, n_val in enumerate(refractive_indices):
    print(f"n_{idx+1} = {n_val.evalf()}")
```
