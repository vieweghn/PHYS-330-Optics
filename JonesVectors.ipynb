```python
import numpy as np

def get_alpha(A,B,delta):
  return 0.5*np.atan2(2*A*B*np.cos(delta),A**2-B**2)
```

```python
AB = 1.0/np.sqrt(2)

np.degrees(get_alpha(AB, AB, 0))
```

```python
def polarizer(theta):
  return np.array([[np.cos(theta)**2, np.sin(theta)*np.cos(theta)],
                   [np.sin(theta)*np.cos(theta), np.sin(theta)**2]])

def half_wave_plate(theta):
  return np.array([[np.cos(2*theta),np.sin(2*theta)],
                   [np.sin(2*theta),-np.cos(2*theta)]])

def quarter_wave_plate(theta):
  return np.array([[np.cos(theta)**2 + 1j*np.sin(theta)**2,(1-1j)*np.sin(theta)*np.cos(theta)],
                   [(1-1j)*np.sin(theta)*np.cos(theta),np.sin(theta)**2 + 1j*np.cos(theta)**2]])
```

```python
def normalize_jones_vector(jones_vector):
  """
  Normalizes a Jones vector by making its first component real and its magnitude one.

  Args:
    jones_vector (np.ndarray): A 2-element complex numpy array representing the Jones vector.

  Returns:
    np.ndarray: The normalized Jones vector.
  """
  # Ensure the input is a numpy array of complex numbers
  jones_vector = np.asarray(jones_vector, dtype=complex)

  # Handle zero vector case early to avoid division by zero
  if np.all(jones_vector == 0):
    return np.asarray([0, 0], dtype=complex)

  # Step 1: Make the first component real by factoring out its phase
  # If the first component is non-zero, calculate and apply its phase factor
  if jones_vector[0] != 0:
    phase_of_first_component = np.angle(jones_vector[0])
    phase_factor = np.exp(-1j * phase_of_first_component)
    vector_with_real_first_component = jones_vector * phase_factor
  else:
    # If the first component is zero, it's already real. No phase rotation needed.
    vector_with_real_first_component = jones_vector

  # Step 2: Normalize the vector to have a magnitude (dot product with complex conjugate) of 1
  # The magnitude is the Euclidean norm for complex vectors.
  magnitude = np.linalg.norm(vector_with_real_first_component)

  # Divide by magnitude to normalize
  normalized_vector = vector_with_real_first_component / magnitude

  # Step 3: Chop values close to zero to exactly zero for cleaner output
  # Apply chopping to real and imaginary parts separately
  normalized_vector.real[np.isclose(normalized_vector.real, 0)] = 0
  normalized_vector.imag[np.isclose(normalized_vector.imag, 0)] = 0

  return normalized_vector
```

```python
normalize_jones_vector(quarter_wave_plate(np.pi/4)@np.array([1,0]))
```

```python
normalize_jones_vector(half_wave_plate(np.pi/4)@np.array([1,0]))
```
