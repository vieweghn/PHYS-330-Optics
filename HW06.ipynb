P Prism: Find the index of refraction of the equilateral prism for the three hydrogen lines we measured in the lab.

image.png

```python
import numpy as np
```

```python
# using Snell's law n1 sin o1 = n2 sin o2, gets to minimum deviation of a prism formula:

def minDeviation_n2(n1, dtheta):
  return n1 * (np.sin(0.5*(np.deg2rad(60) + np.deg2rad(dtheta))) / np.sin(.5 * np.deg2rad(60)))

n1 = 1
theta1 = 179 + 50/60
theta2 = {'Red': 141 + 10/60, 'Aqua': 140 + 22/60, 'Violet': 139 + 56/60}
n2 = []

print('Index of Refraction')
for color, angle in theta2.items():
  n = minDeviation_n2(n1, theta1 - angle)
  print(f'{color}: {n:.4f}')
```
