```python
import numpy as np

def n_plus(A, B, C):
  return np.sqrt((B + np.sqrt(B**2 - 4*A*C)) / (2*A))

def n_minus(A, B, C):
  return np.sqrt((B - np.sqrt(B**2 - 4*A*C)) / (2*A))

def A(u, n):
  return u[0]**2 * n[0]**2 + u[1]**2 * n[1]**2 + u[2]**2 * n[2]**2

def B(u, n):
  return u[0]**2 * n[0]**2 * (n[1]**2 + n[2]**2) + u[1]**2 * n[1]**2 *(n[0]**2 + n[2]**2) + u[2]**2 * n[2]**2 * (n[0]**2 + n[1]**2)

def C(n):
  return n[0]**2 * n[1]**2 * n[2]**2

u_mag = np.sqrt(14)
u = [1, 2, 3] / u_mag

n = [1.5, 1.6, 1.7]

print(f'n = {n_plus(A(u, n), B(u, n), C(n)):.2f}, {n_minus(A(u, n), B(u, n), C(n)):.2f}')

```
