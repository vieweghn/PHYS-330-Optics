```python
def rp (n1, n2):
    return (n1 - n2) / (n1 + n2)
def tp (n1, n2):
    2 * n1 / (n1 + n2)
def o2(o1, n1, n2):
    return np.asin((n1 / n2) * np.sin(o1))

n1 = 1
n2 = 1.5
oi = np.pi / 8

Rag = (rp(n1, n2))**2
T = 1 - Rag # T at g-a, a-g

ot = o2(oi, n2, n1)
R = (np.tan(oi - ot) / np.tan(oi + ot))**2

I = T * R * R * T
print(f'Intensity: {I*1e2:.3f}%')
```

```python
# air to water
n1 = 1
n2 = 1.33
oi = np.linspace(0, np.pi/2, 1000)
def ot(n1, n2, o1):
    return np.asin((n1 / n2) * np.sin(o1))

rs = np.sin(ot(n1, n2, oi) - oi) / np.sin(ot(n1, n2, oi) + oi)
Rs = rs**2
Ts = 1 - Rs

rp = np.tan(ot(n1, n2, oi) - oi) / np.tan(ot(n1, n2, oi) + oi)
Rp = rp**2
Tp = 1 - Rp

plt.plot(np.degrees(oi), Ts, label='Ts')
plt.plot(np.degrees(oi), Tp, label='Tp')
plt.legend()
plt.xlabel('Î¸i')
plt.title('Air to Water')
plt.grid()
plt.show()
```
