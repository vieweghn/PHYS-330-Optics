```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Phi = np.linspace(0,3*np.pi,10000)
Flist = [0.1, 1, 10, 100]

def T(phi, F):
  return 1/(1 + F*np.sin(phi/2)**2)

for F in Flist:
  plt.plot(Phi, T(Phi, F), label=f'F={F}')

plt.legend()
plt.grid()
plt.title(r"Relative transmissivity vs $\Phi$")
plt.xlabel(r"$\Phi$")
plt.ylabel(r"$T(\Phi)$")
plt.show()
```

```python
def calcF(R):
  return 4*R/(1-R)**2

rVec = np.linspace(0,.98,100)
fVec = calcF(rVec)

plt.plot(rVec, fVec)
```
